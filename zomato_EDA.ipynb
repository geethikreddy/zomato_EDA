```python
df_country=pd.read_excel(r'C:\Users\geeth\Downloads\Zomatodataset\Country-Code.xlsx')
df_country.head()
```

```python
# Merging two different dataframes

final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
# checking the type of the elements of the column 

final_df.dtypes
```

```python
# To know what countries use zomato

country_names=final_df['Country'].value_counts().index
country_names
```

```python
# To know which country restaurants use  zomato

countries_with_restaurant=final_df['Country'].value_counts()
countries_with_restaurant
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Ratings Count'})
```

```python
ratings
```

```python
# Find the country names which has been given 0 rating

final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

```python
# To find out which currency is used by which country

final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
# which countries do have online delivery option

final_df[final_df['Has Online delivery']=='Yes'].groupby(['Has Online delivery','Country']).size().reset_index()
```
